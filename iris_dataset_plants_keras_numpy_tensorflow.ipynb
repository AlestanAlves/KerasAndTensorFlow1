```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder
```

```python
uri = 'https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/639388c2cbc2120a14dcf466e85730eb8be498bb/iris.csv'
df = pd.read_csv(uri)

dataset = df.values

x = dataset[:, 0:4].astype('float')
y = dataset[:, 4]

encoder = LabelEncoder()
encoded = encoder.fit_transform(y)

Y2 = np_utils.to_categorical(encoded)

model = Sequential()
model.add(Dense(8,input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['acc']) #user categorical_crosstropy quando tiver softmax na ultima camada
model.fit(x, Y2, epochs=1000)

```
